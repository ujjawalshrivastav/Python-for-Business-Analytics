```python
#Week sales 
sales = []
for day in range(1, 7):
    value = int(input(f"day  {day}:"))
    sales.append(value)

total = sum(sales)
print("This week sales:", total)
```

```python
#While loop 
count = 1
while count <= 5:
    print(count)
    count += 2
```

```python
#While loop 
count = 5
while count <= 50:
    value = input(" ")
    print(count)
    count += 5
```

```python
#for loop example 
print("Sales data for 5 days") 
total_sales = 0

for day in range(1, 6):
    sales = int(input(f"Enter sales for day {day}:"))
    total_sales = total_sales + sales
print("Total Sales: ", total_sales)
```

```python
print("\nCustomer count using while loop")

count = 1
while count <= 3: 
    customer_name = input("Enter customer name: ")
    print("Customer: ", customer_name)
    count += 1
```
