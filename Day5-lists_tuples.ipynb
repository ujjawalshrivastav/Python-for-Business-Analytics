```python
#List
sales  = [1200, 1500, 1800, 2000]
print(sales)
#for one element
print(sales[0])
print(sales[2])
print(sales[-1])


#for multiple elements
print(sales[1:3])
print(sales[:3])
print(sales[3:])
```

```python
#modifying the element 
sales[1] = 1600
print(sales)

#Add element 
print(sales)
sales.append(2200)
print(sales)

sales.insert(1, 1400)

```

```python
#Remove element 
sales.remove(1600)
print(sales)

sales.pop()
print(sales)
```

```python
#Loop with list 

total = 0
for amount in sales: 
    total += amount

print("Total sales: ", total)
```

```python
prices = []

for i in range(5): 
    price = int(input("Enter product price: "))
    prices.append(price)

print("Prices:", prices)
print("Max price:", max(prices))
print("Min price:", min(prices))
print("Average price:", sum(prices) / len(prices))
```
